```python
figure,axis = plt.subplots(1,4,figsize=(12,12))

image = cv.imread('images/00e77bde04dded4f9cc688a18d1f90c7ac4c2be4.tif')
image_blurred = cv.GaussianBlur(image, (0, 0), 3)
image_color = cv.cvtColor(image,cv.COLOR_BGR2HSV)
image_edges = cv.Canny(image_color,30,200)
image_sharp = cv.addWeighted(image, 1.5, image_blurred, -0.5, 0)
axis[0].imshow(image_edges)
axis[1].imshow(image_color)
axis[2].imshow(image_blurred)
axis[3].imshow(image_sharp)
```

```python
image_color = cv.cvtColor(image,cv.COLOR_BGR2HSV)
plt.title('POSITIVE')
plt.imshow(image_color)
plt.show
```
